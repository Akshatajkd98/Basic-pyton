# Exercise 1 :Prime number

```python
num=int(input("enter the number:"))
for i in range (2,num):
    if num%i==0:
        print("it is not prime number")
        break
    else:
        print("it is a prime number")
```

# Exercise 4: Hospital Billing System

```python
room_charge=2000
doctor_consulation_fee=1500
lab_test=300
discount_rate=0.10
medicine_charge=2400
day_stayed=4
number_of_lab_test=3

total_room_cost=day_stayed*room_charge
total_lab_cost=number_of_lab_test*lab_test
total_bill=total_room_cost+total_lab_cost+doctor_consulation_fee+medicine_charge
print("Total_bill:",total_bill)
if total_bill>10000:
    discount=total_bill*discount_rate
    print ("discount amount:",discount)
    final_bill=total_bill-discount
    print("Total amount to be paid:",final_bill)
     
    
   

```
